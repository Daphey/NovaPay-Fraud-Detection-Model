### Build the Machine Learning Model

```python
from sklearn.metrics import (
    roc_auc_score,
    auc,
    classification_report,
    confusion_matrix,
    precision_recall_curve,
    average_precision_score
)
import numpy as np
```

```python
def train_evaluate_model(
    model,
    X_train,
    y_train,
    X_test,
    y_test,
    model_name=""
):
    """
    Trains and evaluates a classification model for fraud detection.
    """
    
    # Train model
    model.fit(X_train, y_train)
    
    # Predictions
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]
    
    # Metrics
    roc_auc = roc_auc_score(y_test, y_proba)
    pr_auc = average_precision_score(y_test, y_proba)
    
    print(f"\n📌 Model: {model_name}")
    print(f"ROC-AUC: {roc_auc:.4f}")
    print(f"PR-AUC: {pr_auc:.4f}")
    
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    
    return {
        "model": model_name,
        "roc_auc": roc_auc,
        "pr_auc": pr_auc
    }
```

```python
def train_evaluate_model(
    model,
    X_train_res,
    y_train_res,
    X_test,
    y_test,
    model_name=""
):
    """
    Trains and evaluates a classification model for fraud detection.
    """
    
    # Train model
    model.fit(X_train_res, y_train_res)
    
    # Predictions
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]
    
    # Metrics
    roc_auc = roc_auc_score(y_test, y_proba)
    pr_auc = average_precision_score(y_test, y_proba)
    
    print(f"\n📌 Model: {model_name}")
    print(f"ROC-AUC: {roc_auc:.4f}")
    print(f"PR-AUC: {pr_auc:.4f}")
    
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    
    return {
        "model": model_name,
        "roc_auc": roc_auc,
        "pr_auc": pr_auc
    }
```

```python
import gradio as gr
import pandas as pd
import joblib

# Load model and preprocessor
model = joblib.load("random_forest_fraud_model.pkl")
#preprocessor = joblib.load("preprocessor.pkl")

def predict_fraud(
    amount_usd,
    txn_velocity_1h,
    txn_velocity_24h,
    account_age_days,
    ip_risk_score,
    risk_score_internal,
    channel,
    new_device,
    location_mismatch,
    kyc_tier
):
    # Create input dataframe
    input_df = pd.DataFrame([{
        "amount_usd": amount_usd,
        "txn_velocity_1h": txn_velocity_1h,
        "txn_velocity_24h": txn_velocity_24h,
        "account_age_days": account_age_days,
        "ip_risk_score": ip_risk_score,
        "risk_score_internal": risk_score_internal,
        "channel": channel,
        "new_device": new_device,
        "location_mismatch": location_mismatch,
        "kyc_tier": kyc_tier
    }])

    # Preprocess
    X_processed = preprocessor.transform(input_df)

    # Predict
    fraud_prob = model.predict_proba(X_processed)[0][1]

    label = "⚠️ High Fraud Risk" if fraud_prob > 0.3 else "✅ Low Fraud Risk"

    return f"Fraud Probability: {fraud_prob:.2%}\n{label}"

# Build Gradio Interface
interface = gr.Interface(
    fn=predict_fraud,
    inputs=[
        gr.Number(label="Transaction Amount (USD)", value=100),
        gr.Number(label="Transactions in last 1 hour", value=1),
        gr.Number(label="Transactions in last 24 hours", value=3),
        gr.Number(label="Account Age (days)", value=120),
        gr.Slider(0, 1, value=0.2, label="IP Risk Score"),
        gr.Slider(0, 1, value=0.3, label="Internal Risk Score"),
        gr.Dropdown(["web", "mobile", "api"], label="Channel"),
        gr.Dropdown([True, False], label="New Device"),
        gr.Dropdown([True, False], label="Location Mismatch"),
        gr.Dropdown(["tier_1", "tier_2", "tier_3"], label="KYC Tier")
    ],
    outputs=gr.Textbox(label="Fraud Risk Result"),
    title="🚨 NovaPay Fraud Detection (Quick Test)",
    description="Enter transaction details to get a fraud risk score"
)

interface.launch()
```
